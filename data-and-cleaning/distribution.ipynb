```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into a DataFrame
df = pd.read_csv('cleandata.csv')

# Check the 'Wavelen' column
if 'Wavelen' not in df.columns:
    raise ValueError("Column 'Wavelen' does not exist in the CSV file.")

# Define the colors based on 'Wavelen' ranges
conditions = [
    (df['Wavelen'] < 590),
    (df['Wavelen'] >= 590) & (df['Wavelen'] <= 660),
    (df['Wavelen'] > 660) & (df['Wavelen'] <= 800),
    (df['Wavelen'] > 800)
]

colors = ['green', 'red', 'brown', 'blue']  # Colors for the different conditions

# Assign colors to each sample
df['color'] = 'black'  # default color
for condition, color in zip(conditions, colors):
    df.loc[condition, 'color'] = color

# Plotting
plt.figure(figsize=(10, 6))
for color in colors:
    subset = df[df['color'] == color]
    plt.hist(subset['Wavelen'], bins=30, color=color, alpha=0.5, label=f'Wavelen {color}')

plt.title('Distribution of Wavelengths')
plt.xlabel('Wavelength (Wavelen)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into a DataFrame
df = pd.read_csv('cleandata.csv')

# Check the 'Wavelen' column
if 'Wavelen' not in df.columns:
    raise ValueError("Column 'Wavelen' does not exist in the CSV file.")

# Define the colors based on 'Wavelen' ranges
conditions = [
    (df['Wavelen'] < 590),
    (df['Wavelen'] >= 600) & (df['Wavelen'] <= 660),
    (df['Wavelen'] > 660) & (df['Wavelen'] <= 800),
    (df['Wavelen'] > 800)
]

colors = ['green', 'red', 'brown', 'blue']  # Colors for the different conditions

# Assign colors to each sample
df['color'] = 'black'  # default color
for condition, color in zip(conditions, colors):
    df.loc[condition, 'color'] = color

# Count the instances in each color group
counts = {color: df[df['color'] == color].shape[0] for color in colors}

# Plotting
plt.figure(figsize=(5, 3))
max_wavelen = df['Wavelen'].max()

# Define bins considering the range and the data distribution
bins = range(400, int(max_wavelen) + 10, 10)  # Adjust bin width as necessary

for color in colors:
    subset = df[df['color'] == color]
    # Update label to include the count of instances
    plt.hist(subset['Wavelen'], bins=bins, color=color, alpha=0.5, label=f'{color}: {counts[color]}', density=False)

plt.xlim(400, max_wavelen)
plt.title('Distribution of Wavelengths')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into a DataFrame
df = pd.read_csv('cleandata.csv')

# Check the 'Wavelen' column
if 'Wavelen' not in df.columns:
    raise ValueError("Column 'Wavelen' does not exist in the CSV file.")

# Define the colors based on 'Wavelen' ranges
conditions = [
    (df['Wavelen'] < 590),
    (df['Wavelen'] >= 590) & (df['Wavelen'] <= 660),
    (df['Wavelen'] > 660) & (df['Wavelen'] <= 800),
    (df['Wavelen'] > 800)
]

colors = ['green', 'red', 'brown', 'blue']  # Colors for the different conditions
group_names = ['Green', 'Red', 'Farred', 'NIR']  # Custom names for each group

# Assign colors to each sample
df['color'] = 'black'  # default color
for condition, color in zip(conditions, colors):
    df.loc[condition, 'color'] = color

# Count the instances in each color group and create label text
labels = {
    color: f'{name}: {df[df["color"] == color].shape[0]}' 
    for color, name in zip(colors, group_names)
}

# Plotting
plt.figure(figsize=(5,3))
max_wavelen = df['Wavelen'].max()

# Define bins considering the range and the data distribution
bins = range(400, int(max_wavelen) + 10, 10)  # Adjust bin width as necessary

for color in colors:
    subset = df[df['color'] == color]
    # Update label to include the custom name and the count of instances
    plt.hist(subset['Wavelen'], bins=bins, color=color, alpha=1, label=labels[color], density=False)

plt.xlim(400, max_wavelen)
#plt.title('Distribution of Wavelengths')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
